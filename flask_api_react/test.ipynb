```python
## インポート
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import os
import jpy_datareader.data as web

# e-StatのAPI_KEY
api = os.environ.get("API_KEY")
```

```python
'''
データ元：人口動態調査/人口動態統計/確定数/人口/年次・都道府県・性別人口
https://www.e-stat.go.jp/stat-search/database?page=1&layout=dataset&toukei=00450011&tstat=000001028897&year=20220&hclass=10010&statdisp_id=0003411884&metadata=1&data=1
'''
df = web.DataReader("0003411884", "estat", api_key=api,retry_count=3)
```

```python
map_df = df[(df['時間軸(年次)'] == '2022年') & (df['性別'] == '総数') & (df['都道府県（特別区－指定都市再掲）'] != '全国')]
map_df = map_df.reset_index().drop(['index', '時間軸(年次)', '性別'], axis=1)
map_df.rename(columns={'都道府県（特別区－指定都市再掲）': 'name', '人口': 'population_jp'}, inplace=True)
```

```python
# 都道府県コードを追加するためにマッピングを適用します
map_df['code'] = map_df['name'].map(prefecture_codes)
```
